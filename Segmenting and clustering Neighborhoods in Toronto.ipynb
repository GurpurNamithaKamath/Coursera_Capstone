### Getting the Postal codes,Borough and neighborhoods in a pd by scraping the wikipedia page with Beautiful Soup.

```python
!conda install -c conda-forge wikipedia
```

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
import requests # library to handle requests
```

```python
website_url=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
print(len("My_table"))
```

```python
links=My_table.findAll('tr')
links
```

```python
import pandas as pd

column_names = ['PostalCode','Borough','Neighborhood'] 
neighborhoods = pd.DataFrame(columns=column_names)
```

```python
neighborhoods
```

```python
postcodes = []
boroughs =[]
neighbors = []

for tr in links[1:]:
    tds = tr.find_all('td')
    #print("Postcode: %s, Borough: %s, Neighbor: %s" % (tds[0].text, tds[1].text, tds[2].text))
    postcodes.append(tds[0].text)
    boroughs.append(tds[1].text)
    neighbors.append(tds[2].text)
    
print(postcodes)
print(boroughs)
print(neighbors)
```

```python
for i in range(len(neighbors)):
    neighbors[i] = neighbors[i].replace('\n','')
df = pd.DataFrame()
   
#print(neighbors[0])

# three columns: PostalCode, Borough, and Neighborhood
df['PostalCode'] = postcodes
df['Borough'] = boroughs
df['Neighborhood'] = neighbors

df
```

```python
df = df[df.Borough != 'Not assigned'] # removed the rows where Borough is NOT ASSIGNED.ignoring cells that have not been assigned.
df
df.reset_index(drop=True)# assigning index from zero
```

```python
#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. 
#So for the 9th cell in the table on the Wikipedia page, the value of the Borough and the Neighborhood columns will be Queen's Park.
df[df.Neighborhood == 'Not assigned']
df.Neighborhood[9]= "Queen's Park"

df

df.reset_index(drop=True)
```

```python
d2  = df.groupby(['PostalCode', 'Borough'])["Neighborhood"].agg(lambda col: ', '.join(col)).reset_index()##grouping the data by postalcode and borough

d2
```

```python
'The number of rows of the dataframe is {}'.format(d2.shape[0])
```

```python
print('The dataframe has {} boroughs and {} Neighborhood.'.format(
        len(d2['Borough'].unique()),
        d2.shape[0]
  )
     )      
```

## using the geocoder package for creating a dataframe with longitude and latitudes.

```python
from geopy.geocoders import Nominatim
```

```python
data = pd.read_csv("http://cocl.us/Geospatial_data") 
```

```python
data.head()
```

```python
data = data[['Latitude','Longitude']]
postal_code_canada_df = pd.concat([d2, data], axis=1)

postal_code_canada_df
```

## Getting the map 

```python
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
!conda install -c conda-forge folium=0.5.0
import folium # map rendering library
```

```python
address='Toronto,Canada'
geolocator = Nominatim(user_agent="Canada_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
for lat, lng, label in zip(
postal_code_canada_df['Latitude'],
postal_code_canada_df['Longitude'],    
postal_code_canada_df['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
